```python
import json
import pandas as pd 
import re
```

```python
with open('/opt/ml/mrc-level2-nlp-07/code/outputs/mixing_bowl/no1.json', 'r') as f:
    json_data1 = json.load(f)
print(json_data1['mrc-0-000114'])
```

```python
with open('/opt/ml/mrc-level2-nlp-07/code/outputs/mixing_bowl/no2.json', 'r') as f:
    json_data2 = json.load(f)
print(json_data2['mrc-0-000114'])
```

```python
for k in json_data1.keys():
    if json_data1[k] != json_data2[k]:
        print(k, json_data1[k], '|', json_data2[k])
```

```python
for k in json_data1.keys():
    check = 0
    data1 = json_data1[k]
    data2 = json_data2[k]

    if data1 in data2 and len(data1) <= len(data2):
        check = 1
        print("working too")
    elif data2 in data1 and len(data1) >= len(data2):
        check = 2
        print("working")
    if check == 1:
        json_data1[k] = data1
    elif check == 2:
        json_data1[k] = data2
    json_data1[k] = json_data1[k].split(r"\\n")
    json_data1[k] = json_data1[k][-1].strip()
    json_data1[k] = ' '.join(json_data1[k].split())

    
```

```python
print(json_data1['mrc-0-004432'])
```

```python
with open("/opt/ml/code/mixing_bowl/result.json",'w') as make_file:
    json.dump(json_data1,make_file,indent="\t",ensure_ascii=False)
```
